```python
# @title Texto de título padrão
# Importação de bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import (accuracy_score, confusion_matrix,
                            classification_report, precision_score,
                            recall_score, f1_score)

# 1. Carregamento e Preparação dos Dados
print("1. Carregando e preparando os dados...")

# Carrega o dataset (substitua pelo caminho do seu arquivo)
df = pd.read_csv('spotify.csv')

# Pré-processamento dos dados
# Vamos selecionar algumas colunas relevantes e simplificar os gêneros
df = df[['danceability', 'energy', 'valence', 'popularity', 'track_genre']]

# Vamos agrupar os gêneros em categorias mais amplas para simplificar a classificação
def simplify_genre(genre):
    genre = genre.lower()
    if 'pop' in genre:
        return 'pop'
    elif 'rock' in genre:
        return 'rock'
    elif 'hip' in genre or 'rap' in genre:
        return 'hiphop'
    elif 'jazz' in genre:
        return 'jazz'
    elif 'classical' in genre:
        return 'classical'
    else:
        return 'other'

df['genre_group'] = df['track_genre'].apply(simplify_genre)

# Remove gêneros com poucas amostras (para balanceamento)
genre_counts = df['genre_group'].value_counts()
main_genres = genre_counts[genre_counts > 500].index
df = df[df['genre_group'].isin(main_genres)]

# Codifica os gêneros para valores numéricos
le = LabelEncoder()
df['genre_encoded'] = le.fit_transform(df['genre_group'])

# Separa features (X) e target (y)
X = df[['danceability', 'energy', 'valence', 'popularity']]
y = df['genre_encoded']

# Divide em conjuntos de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Normaliza os dados
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# 2. Implementação dos Algoritmos
print("\n2. Implementando os algoritmos...")

# Dicionário para armazenar os modelos e resultados
models = {
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "SVM": SVC(random_state=42),
    "Naive Bayes": GaussianNB()
}

# Treina os modelos e armazena os resultados
results = []
for name, model in models.items():
    print(f"  - Treinando {name}...")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Calcula métricas
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')

    # Armazena resultados
    results.append({
        'Model': name,
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1 Score': f1
    })

    # Exibe relatório de classificação
    print(f"\n{classification_report(y_test, y_pred, target_names=le.classes_)}")

    # Plota matriz de confusão
    plt.figure(figsize=(8, 6))
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=le.classes_, yticklabels=le.classes_)
    plt.title(f'Matriz de Confusão - {name}')
    plt.ylabel('Verdadeiro')
    plt.xlabel('Predito')
    plt.show()

# 3. Comparação dos Modelos
print("\n3. Comparando os modelos...")

# Cria dataframe com os resultados
results_df = pd.DataFrame(results)

# Exibe tabela comparativa
print("\nTabela Comparativa:")
print(results_df)

# Gráfico de comparação de acurácia
plt.figure(figsize=(10, 6))
sns.barplot(x='Model', y='Accuracy', data=results_df, palette='viridis')
plt.title('Comparação de Acurácia dos Modelos')
plt.ylim(0, 1)
plt.ylabel('Acurácia')
plt.xlabel('Modelo')
plt.show()

# Gráfico de comparação das métricas
metrics_df = results_df.melt(id_vars='Model', var_name='Metric', value_name='Value')
plt.figure(figsize=(12, 6))
sns.barplot(x='Model', y='Value', hue='Metric', data=metrics_df, palette='mako')
plt.title('Comparação das Métricas dos Modelos')
plt.ylabel('Valor')
plt.xlabel('Modelo')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# 4. Discussão dos Resultados
print("\n4. Discussão dos Resultados:")
best_model = results_df.loc[results_df['Accuracy'].idxmax()]
print(f"\nO melhor modelo foi {best_model['Model']} com uma acurácia de {best_model['Accuracy']:.2f}.")
print("\nPossíveis razões para o desempenho:")
print("- Random Forest geralmente tem bom desempenho em diversos tipos de dados")
print("- SVM pode ter desempenho inferior com hiperparâmetros padrão")
print("- Naive Bayes assume independência entre features, o que pode não ser verdade aqui")
print("- Árvores de decisão podem sofrer de overfitting sem ajuste de parâmetros")
```
